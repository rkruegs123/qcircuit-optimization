```python
import pyzx as zx
import matplotlib.pyplot as plt
from pyzx.utils import VertexType
import congruences
```

```python
c = zx.generate.CNOT_HAD_PHASE_circuit(qubits=5, depth=10, clifford=False)
zx.draw(c, labels=True)
```

```python
g = c.to_graph()

zx.full_reduce(g) # How can we get graph state?
# g.normalize()
zx.draw(g, labels=True)
ts = g.types()
print(ts)
print(ts[0] == VertexType.BOUNDARY)
```

```python
congruences.unfuse(g, 6)
zx.draw(g, labels=True)
print([(v, congruences.is_gadget(g, v)) for v in g.vertices()])
```
